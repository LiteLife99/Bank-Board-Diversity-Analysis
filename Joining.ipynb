```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline
import seaborn as sns
```

```python
#converting text data to dataframe
df = pd.read_csv('prowess_bank_data.txt', sep='\t', header=None)
df.head()
```

```python
#Correcting the column names
df.rename(columns=df.iloc[0], inplace = True)
df.drop([0], inplace = True)
df.head()
```

```python
df
```

```python
df['Year'] = df['sa_finance1_year'].str.slice(0,4)
df['Year']
```

```python
df['Year'].describe
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['Year'].describe
```

```python
df_2015 = df[df['Year']==2015]
df_2016 = df[df['Year']==2016]
df_2017 = df[df['Year']==2017]
df_2018 = df[df['Year']==2018]
df_2019 = df[df['Year']==2019]
```

```python
df_2019
```

```python
df_2015.rename(columns={'sa_company_name': 'Company'}, inplace=True)
df_2016.rename(columns={'sa_company_name': 'Company'}, inplace=True)
df_2017.rename(columns={'sa_company_name': 'Company'}, inplace=True)
df_2018.rename(columns={'sa_company_name': 'Company'}, inplace=True)
df_2019.rename(columns={'sa_company_name': 'Company'}, inplace=True)
```

```python
df_index_2019 = pd.read_excel('index_data_2019.xlsx', index_col=0)
df_index_2019.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'], axis=1, inplace=True)
df_index_2019=df_index_2019.reset_index()
df_index_2019
```

```python
df_index_2018 = pd.read_excel('index_data_2018.xlsx', index_col=0)
df_index_2018.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11','Unnamed: 12'], axis=1, inplace=True)
df_index_2018=df_index_2018.reset_index()

df_index_2017 = pd.read_excel('index_data_2017.xlsx', index_col=0)
df_index_2017.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11','Unnamed: 12'], axis=1, inplace=True)
df_index_2017=df_index_2017.reset_index()

df_index_2016 = pd.read_excel('index_data_2016.xlsx', index_col=0)
df_index_2016.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11'], axis=1, inplace=True)
df_index_2016=df_index_2016.reset_index()

df_index_2015 = pd.read_excel('index_data_2015.xlsx', index_col=0)
df_index_2015.drop(['Unnamed: 9', 'Unnamed: 10','Unnamed: 11'], axis=1, inplace=True)
df_index_2015=df_index_2015.reset_index()
```

```python
df_index_2015
```

```python
df_index_2015.rename(columns={'company': 'Company'}, inplace=True)
df_index_2016.rename(columns={'company': 'Company'}, inplace=True)
df_index_2017.rename(columns={'company': 'Company'}, inplace=True)
df_index_2018.rename(columns={'company': 'Company'}, inplace=True)
```

```python
df_index_2018
```

```python
df_2016
```

```python
findata_2019 = pd.merge(df_index_2019, df_2019, on='Company')
findata_2018 = pd.merge(df_index_2018, df_2018, on='Company')
findata_2017 = pd.merge(df_index_2017, df_2017, on='Company')
findata_2016 = pd.merge(df_index_2016, df_2016, on='Company')
findata_2015 = pd.merge(df_index_2015, df_2015, on='Company')
findata_2019
```
