```python
df2.drop(df2.index[len(df2)-1],inplace=True)
# df2['Tenure'].astype(int)
# df2['Tenure'].mean()
```

```python
df2.head()
```

```python
# contained some other garbage, so null check was not enough
df2['Tenure'] = pd.to_numeric(df2['Tenure'], errors='coerce')

df3 = df2.dropna(subset=['Tenure'])

# final conversion now worked
df3['Tenure'] = df3['Tenure'].astype(int)

print(df3['Tenure'].mean())
```

```python
df2['Age'] = pd.to_numeric(df2['Age'], errors='coerce')

df2['Age'] = df2['Age'].replace(0,np.nan)

df4 = df2.dropna(subset=['Age'])

# final conversion now worked
df4['Age'] = df4['Age'].astype(int)

print(df4['Age'].mean())
```
