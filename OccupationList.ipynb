```python
#importing some libraries
import pandas as pd
import matplotlib.pyplot as ptl
%matplotlib inline
import seaborn as sns
```

```python
#reading the file
df = pd.read_excel('IndianBoardsData2019.xlsx', index_col=0)
df
```

```python
df.columns
```

```python
x = df['Education1']
y = df['Education2']
z = df['Education3']
from itertools import chain
edulist = list(chain(x,y,z))
edulist
#print(len(edulist))
```

```python
#dropping na from list
#reslist = list(filter(None, edulist))
cleanedList = [x for x in edulist if str(x) != 'nan']
print(len(cleanedList))
cleanedList
```

```python
#getting unique values in list
uniquelist = list(set(cleanedList))
print(len(uniquelist))
uniquelist
```

```python
#string matching 
matching = [s for s in uniquelist if "C.A" in s]
matching
```
